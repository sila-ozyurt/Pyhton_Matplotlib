```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1=np.linspace(0,10,20)
arr2=arr1**2
```

```python
fig=plt.figure()
axes=fig.add_axes([0.6,0.6,0.8,0.8])
#0.8 ler sırasıyla width ve height
axes.plot(arr1,arr2,"r--")
axes.set_xlabel("x axes")
axes.set_ylabel("y axes")
axes.set_title("main title")
plt.show()



#figure kullanarak daha kolay özelleştirebileceğimiz grafikler çizeriz



# plt.plot(arr1,arr2,"y*-")
# plt.xlabel("x axes")
# plt.ylabel("y axes")
# plt.title("main title")
# plt.show()





# plt.subplot(2,1,1)
# #2 sıra olacak 1 kolon 1.grafiği çiziyorum
# plt.plot(arr1,arr2,"r--")
# plt.show()


# plt.subplot(2,1,2)
# #2 sıra olacak 1 colon olacak 2.grafiği çiziyorum
# plt.plot(arr2,arr1,"o")
# plt.show()
```
