```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1=np.random.random(10)
arr2=np.random.randn(10)
```

```python
fig=plt.figure(figsize=(7,7),dpi=100)
axes=fig.add_axes([0.1,0.1,0.7,0.7])
axes.plot(arr1,arr2,"r",label="arr1/arr2")
axes.plot(arr2,arr2,"b*-",label="arr2/arr2")
axes.legend(loc=2)
plt.show()
```

```python
(figure1,axeses)=plt.subplots()
axeses.plot(arr1,arr2,"r",label="new axes",alpha=0.5)
#rgb to hex kodu bulup baka renklerde de graphic izdirebiliriz
#color="hex kod" eklinde yaparz
#alpha=1 iken tam, azladka transparan

axeses.plot(arr1,arr1,"y",label="arr1/arr1",alpha=0.3)
axeses.legend(loc=4)
plt.show()
```

```python
(figure2,newAxeses)=plt.subplots(figsize=(7,7),dpi=70)
newAxeses.plot(arr1,arr2,"b",label="arr1/arr2",alpha=1,linewidth=4)
newAxeses.plot(arr2,arr1,"r--",label="arr2,arr1",alpha=0.5,linewidth=8,linestyle="dashdot")
newAxeses.plot(arr1,arr1,"b",label="arr1/arr1",alpha=1,linewidth=3,linestyle="dashed")
newAxeses.plot(arr2,arr2,"y",label="arr2/arr2",alpha=0.2,linewidth=5,linestyle="dotted",marker="o",markersize=9,markerfacecolor="b")
newAxeses.legend(loc=2)
# linestyle    : - -- None  dashdot dotted dashed
plt.show()

```
