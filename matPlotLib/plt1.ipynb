```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ageList=[10,30,40,50,45,56,67,78,34,23]
weightList=[20,60,87,86,90,97,54,22,34,66]

ageNum=np.array(ageList)
weightNum=np.array(weightList)
```

```python
plt.plot(ageNum,weightNum,"g--")
plt.xlabel("ages")
plt.ylabel("weights")
plt.title("graphic")
plt.show()
#ilk verdiğim liste x eksenine yazıldı
#ikinci liste y ekseninde
# g dediğin de green, r red, b blue, y yellow
```

```python
# #0 ile 10 arasında 20 tane degisken olustur
arr1=np.linspace(0,10,20)
arr1
```

```python
arr2=arr1**3
arr2
```

```python
plt.plot(arr1,arr2,"y*-")
plt.xlabel("arr")
plt.ylabel("arr**3")
plt.title("graphic")
plt.show()
```
