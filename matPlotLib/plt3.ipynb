```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1=np.linspace(0,10,20)
arr2=arr1**3
```

```python
fig=plt.figure(figsize=(9,9),dpi=100)
#figsize ile de width ve height belirleyebiliriz
#dpi çözünürlük için kullanılır

axes=fig.add_axes([0.1,0.1,0.9,0.9])
axes.plot(arr1,arr2,"r--",label="arr1/arr2")
#başka bir axes oluşturursan aynı graphic icinde tiny graphic oluşturur
#direkt axes.plot diyip çizdirirsen aynı grapic içinde farklı renkle çizer

axes.plot(arr1,arr1,"o",label="arr1/arr1")
axes.legend(loc=2)
#loc 1 2 3 4 5 vb yaparsın
plt.show()
fig.savefig("myfig.png",dpi=70)

```
